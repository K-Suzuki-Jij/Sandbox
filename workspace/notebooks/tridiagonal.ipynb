```python
import numpy as np
import matplotlib.pyplot as plt
from strum import Strum, tridiag
n =1000
th=1e-14
```

```python
np.random.seed(1)
a = np.random.uniform(0,1,n)
b = np.random.uniform(0,1,n-1)
strum = Strum(th)
```

```python
eig1 = np.linalg.eigvalsh(tridiag(n,a,b))
```

```python
eig2 = strum.bis(n,a,b)
```

```python
plt.plot(np.arange(1, n+1, 1), eig1, label="np")
plt.plot(np.arange(1, n+1, 1), eig2, label="my")
plt.legend()
```

```python
plt.plot(np.arange(1, n+1, 1), eig2-eig1, label="my")
plt.xlabel("n")
plt.ylabel("Absolute errors of n-th eigenvalue")
```
